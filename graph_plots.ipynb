```python
import matplotlib.pyplot as plt
```

```python
from functools import singledispatchmethod

# batch size versus flops
batches = [8,16,32, 64, 128, 256]
time = [416045260,772183859,1572480000, 2934992486, 6006640640, 12013281280]

plt.figure(figsize=(10,6))
plt.plot(batches, time, label='flops', color='blue', marker='o')

for i, (x, y) in enumerate(zip(batches, time)):
    plt.annotate(f'({x}, {y})', (x, y), textcoords="offset points", xytext=(2,3), ha='center', fontsize='6')


plt.xlabel('batch size')
plt.ylabel('flops')

# Display the plot
plt.show()
```

```python
from functools import singledispatchmethod

# estimated time vs measured time
estimated_time = [156800000,313600000,627200000, 1254400000, 2508800000, 5017600000]
measured_time = [416045260,772183859,1572480000, 2934992486, 6006640640, 12013281280]

slope = (772183859-416045260) / (313600000-156800000)

estimated_time_in_millions = [e/1e6 for e in estimated_time]
measured_time_in_millions = [m/1e6 for m in measured_time]

plt.figure(figsize=(10,6))

# Create a plot
plt.plot(estimated_time_in_millions, measured_time_in_millions, label='flops', color='blue', marker='o')

for i, (x, y) in enumerate(zip(estimated_time_in_millions, measured_time_in_millions)):
    plt.annotate(f'({x}, {y:0.2f})', (x, y), textcoords="offset points", xytext=(2,3), ha='left', fontsize='6')


plt.xlabel('estimated_time_in_millions')
plt.ylabel('measured_time_in_millions')

# Display the plot
plt.show()

print("Slope of the graph is approximately",slope)
```
